# Linear Regression using multiple features

```python
x = weather_df[['Humidity' , 'Visibility' ,'Wind_Bearing' ,'Pressure']]
y = weather_df.Temperature
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
def initialize_weights():
    return np.random.randn(1, 1), np.random.randn()  # Random initialization

# Step 3: Implement the LMS algorithm
def lms(X, y, learning_rate, epochs):
    w, b = initialize_weights()
    n_samples = X.shape[0]
    mse_history = []
    
    for epoch in range(epochs):
        for i in range(n_samples):
            y_pred = w * X[i] + b  # Prediction
            error = y[i] - y_pred  # Calculate the error
            w += learning_rate * error * X[i]  # Update weights
            b += learning_rate * error  # Update bias
            
        # Calculate MSE after each epoch
        y_preds = w * X + b
        mse = np.mean((y - y_preds)**2)
        mse_history.append(mse)
    
    return w, b, mse_history
```

```python
learning_rates = [0.01, 0.1, 0.5]
epochs = 50

plt.figure(figsize=(10, 7))

for lr in learning_rates:
    w, b, mse_history = lms(X_train, y_train, lr, epochs)
    plt.plot(mse_history, label=f"Learning rate: {lr}")
    print(f"Learning Rate: {lr}, Final Weights: {w[0][0]:.3f}, Bias: {b:.3f}, Final MSE: {mse_history[-1]:.3f}")

plt.title("MSE vs Epochs for different learning rates")
plt.xlabel("Epochs")
plt.ylabel("Mean Squared Error")
plt.legend()
plt.show()
```
